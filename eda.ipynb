```python
sales = m2[['quantity','unit','unit_price','total_price','item_name','desc','supplier','date','quarter','month','year']]
sales.head()
```

```python
sales['date'] = pd.to_datetime(sales['date'], format='%d-%m-%Y %H:%M')
sales['date']
```

```python
sales.columns
```

```python
# sales overtime
plt.figure(figsize=(12, 6))
sales.groupby('date')['total_price'].sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# sales by year
plt.figure(figsize=(12, 6))
sales.groupby('year')['total_price'].sum().plot(kind='line')
plt.title('Total Sales by year')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.show()
```

```python
sales[sales['year']==2021]['month'].unique()
# this means data for 2021 not complete, would exclude it for analysis 
sales = sales[sales['year']!=2021]

sales[sales['year']==2014]['month'].nunique()
```

```python
# sales by year
plt.figure(figsize=(12, 6))
sales.groupby('year')['total_price'].sum().plot(kind='line')
plt.title('Total Sales by year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

# very rapid growth from 2014-2015
# some drop in sales in 2016, 2019
```

```python
# sales by quarter
plt.figure(figsize=(12, 6))
sales.groupby('quarter')['total_price'].sum().plot(kind='line')
plt.title('Total Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.show()

# high sale in Q3, Q4
```

```python
# sales by month 
plt.figure(figsize=(12, 6))
sales.groupby('month')['total_price'].sum().plot(kind='line')
plt.title('Total Sales by month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()

# low sale in 2, 4, 6, 9, 11 
```

```python
sales['year_month'] = sales['year'].astype(str) + '-' + sales['month'].astype(str)
sales['year_month']
sales['year_month'] = pd.to_datetime(sales['year_month'], format='%Y-%m')
```

```python
# sales by year-month 
plt.figure(figsize=(12, 6))
sales.groupby('year_month')['total_price'].sum().plot(kind='line')

plt.title('Total Sales by month')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.show()

# can do more eda later, but it looks like for each year, the seasonal pattern is pretty consistent 
```

```python
sales['quarter']
sales['quarter_int'] = sales['quarter'].str.extract('(\d)').astype(int)
sales['quarter_int'] 
```

```python
# sales['year_month'] = sales['year'].astype(str) + '-' + sales['quarter'].astype(str)
# sales['year_month']
# sales['year_month'] = pd.to_datetime(sales['year_month'], format='%Y-%m')

# Create 'year_quarter' column
sales['year_quarter'] = pd.PeriodIndex(year=sales['year'], quarter=sales['quarter_int'], freq='Q')
```

```python
# sales by year-quarter
plt.figure(figsize=(12, 6))
sales.groupby('year_quarter')['total_price'].sum().plot(kind='line')

plt.title('Total Sales by month')
plt.xlabel('Year-Quarter')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group by the 'year_quarter' and sum 'total_price'
quarterly_sales = sales.groupby('year_quarter')['total_price'].sum()

# Convert PeriodIndex to timestamps for plotting
quarterly_sales.index = quarterly_sales.index.to_timestamp()

plt.figure(figsize=(12, 6))

# Define a colormap
colors = plt.cm.tab10(np.linspace(0, 1, 4))  # Using tab10 colormap for 4 quarters

# Plot the line
plt.plot(quarterly_sales.index, quarterly_sales, marker='', color='grey', linewidth=1)

# Scatter plot for each quarter with different colors
for i, quarter in enumerate(range(1, 5)):
    quarter_sales = quarterly_sales[quarterly_sales.index.quarter == quarter]
    plt.scatter(quarter_sales.index, quarter_sales, color=colors[i], label=f'Q{quarter}')

# Set custom x-ticks
plt.xticks(ticks=quarterly_sales.index, labels=quarterly_sales.index.to_period('Q').astype(str), rotation=45)

plt.title('Total Sales by Quarter')
plt.xlabel('Year-Quarter')
plt.ylabel('Total Sales')
plt.legend(title='Quarter')
plt.show()
```

```python
# top selling products 

# Calculate total sales per product
total_sales_per_product =sales.groupby('item_name')['total_price'].sum().reset_index()

# Plot top 10 best-selling products
top_products = total_sales_per_product.sort_values(by='total_price', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_products, x='total_price', y='item_name')
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.title('Top 10 Best-Selling Products')
plt.show()

# further: for each selling product, what category belongs to ? 
```

```python
# top selling categories

# Calculate total sales per product
total_sales_per_category =sales.groupby('desc')['total_price'].sum().reset_index()

# Plot top 10 best-selling products
top_categories = total_sales_per_category.sort_values(by='total_price', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_categories, x='total_price', y='desc')
plt.xlabel('Total Sales')
plt.ylabel('Categories')
plt.title('Top 10 Best-Selling categories')
plt.show()
```

```python
# Sales Analysis
# Merge fact_df with time_df to get dates
sales_df = fact_df.merge(time_df, on='time_key')

# Aggregate total sales per month
monthly_sales = sales_df.groupby('month')['total_price'].sum().reset_index()

# Plot total sales over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='total_price')
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.show()
```

```python
# customer analysis
# Merge fact_df with customers_df to get customer information
customer_sales_df = fact_df.merge(customers_df, on='coustomer_key')

# Calculate average spend per customer
avg_spend_per_customer = customer_sales_df.groupby('coustomer_key')['total_price'].mean().reset_index()
avg_spend_per_customer.columns = ['coustomer_key', 'avg_spend']

# Plot distribution of average spend
plt.figure(figsize=(12, 6))
sns.histplot(avg_spend_per_customer['avg_spend'], bins=30, kde=True)
plt.xlabel('Average Spend')
plt.ylabel('Frequency')
plt.title('Distribution of Average Spend per Customer')
plt.show()
```
